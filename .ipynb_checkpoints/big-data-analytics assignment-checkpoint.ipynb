### 4.0 Filtering The 2020 Data for Accident , Vehicle and Casualty

#### 4.0.1 Create a Function to query data for the year 2020

```python
#write a function that filters by accident_year which is 2020
def filter2020(data):
    try:
        info = pd.read_sql_query(f"SELECT * FROM {data} WHERE accident_year = 2020", conn)
        return info
    except():
        print("data not available")
```

#### 4.1 Query the Accident data for the year 2020

```python
# Run the function that queries accident data for 2020
accident2020 = filter2020("accident")
accident2020
```

#### Are there significant hours of the day and days of the week on which accident occurs?


```python
# trying to convert time to datetime
accident2020['datetime'] = pd.to_datetime(accident2020['date'] + ' ' + accident2020['time'] , dayfirst=True)

# transform to hour and day of the week.
accident2020['hour'] = accident2020['datetime'].dt.hour
accident2020['day_of_week'] = accident2020['datetime'].dt.day_name()

# Visualization

# Distribution of accidents by hour of the day
plt.figure(figsize=(16, 6))
accident2020['hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.grid(False)
plt.show()

# Distribution of accidents by day of the week
plt.figure(figsize=(16, 6))
accident2020['day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).plot(kind='bar')
plt.title('Distribution of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.grid(False)
plt.show()
```
